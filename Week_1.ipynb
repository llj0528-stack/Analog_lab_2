- Assignment 3: f<sub>T</sub>
  - Derive 
  $$f_T = \frac{g_m}{2\pi c_gg}$$
  - Plot f<sub>T</sub> - g<sub>m</sub>/I<sub>D</sub> and A<sub>V</sub> - g<sub>m</sub>/I<sub>D</sub> under different channel length

```python
L_set = np.array([0.05, 0.2, 0.5, 1])
gm_ID_vec = np.arange(4, 30, 0.1) #[4, 4.1, ... 29.9, 30]
fT_vec = look_up_vs_gm_id(nch, "GM_CGG", gm_ID_vec, vds = 0.55, vsb = 0, l = L_set)/(2*np.pi) # gm_ID is a function of Vgs
Av_vec = look_up_vs_gm_id(nch, "GM_GDS", gm_ID_vec, vds = 0.55, vsb = 0, l = L_set)

# Plot fT-gm_ID
for i in range(len(L_set)):
    plt.plot(gm_ID_vec, fT_vec[i], label = f"L = {L_set[i]:.2f}um")
plt.xlabel("gm_ID (1/V)")
plt.ylabel("fT_vec (Hz)")
plt.yscale("log")
plt.title('fT-gm_ID under different L')
plt.legend()
plt.show()

# Plot Av-gm_ID
for i in range(len(L_set)):
    plt.plot(gm_ID_vec, Av_vec[i], label = f"L = {L_set[i]:.2f}um")
plt.xlabel("gm_ID (1/V)")
plt.ylabel("Av_vec (V/V)")
plt.title('Av-gm_ID under different , Why different from the lab manual?')
plt.legend()
plt.show()
```

- Conclusion from the former 2 plots
  - L larger -> A<sub>v</sub> bigger, because L larger -> channel length modulation smaller -> r<sub>o</sub> smaller
  - L larger -> f<sub>T</sub> smaller, because L larger -> c<sub>gg</sub> smaller & g<sub>m</sub> larger 
  - gm_ID larger -> A<sub>v</sub> smaller & f<sub>T</sub>

- **g<sub>m</sub>/I<sub>D</sub> systematic design methodology**
  - From design specifications and constrains, determine g<sub>m</sub>
  - Pick L: 
    - Small L -> High f<sub>T</sub>, small Area
    - Large L -> High A<sub>v</sub>, large area
  - Pick g<sub>m</sub>/I<sub>D</sub>: 
    - Large g<sub>m</sub>/I<sub>D</sub> -> Low-power, high A<sub>v</sub>
    - Small g<sub>m</sub>/I<sub>D</sub> -> High f<sub>T</sub>, small Area
  - Derive I<sub>D</sub> by using the result of g<sub>m</sub> and g<sub>m</sub>/I<sub>D</sub>
  - Determine W using I<sub>D</sub>/W denormalization
