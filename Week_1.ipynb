- Assignment 3: f<sub>T</sub>
  - Derive 
  $$f_T = \frac{g_m}{2\pi c_gg}$$
  - Plot f<sub>T</sub> - g<sub>m</sub>/I<sub>D</sub> and A<sub>V</sub> - g<sub>m</sub>/I<sub>D</sub> under different channel length

```python
L_set = np.array([0.05, 0.2, 0.5, 1])
gm_ID_vec = np.arange(4, 30, 0.1) #[4, 4.1, ... 29.9, 30]
fT_vec = look_up_vs_gm_id(nch, "GM_CGG", gm_ID_vec, vds = 0.55, vsb = 0, l = L_set)/(2*np.pi) # gm_ID is a function of Vgs
Av_vec = look_up_vs_gm_id(nch, "GM_GDS", gm_ID_vec, vds = 0.55, vsb = 0, l = L_set)

# Plot fT-gm_ID
for i in range(len(L_set)):
    plt.plot(gm_ID_vec, fT_vec[i]/1e9, label = f"L = {L_set[i]:.2f}um")
plt.xlabel("gm_ID (1/V)")
plt.ylabel("fT (GHz)")
plt.yscale("log")
plt.title('fT-gm_ID under different L')
plt.legend()
plt.show()

# Plot Av-gm_ID
for i in range(len(L_set)):
    plt.plot(gm_ID_vec, Av_vec[i], label = f"L = {L_set[i]:.2f}um")
plt.xlabel("gm_ID (1/V)")
plt.ylabel("Av (V/V)")
plt.title('Av-gm_ID under different')
plt.legend()
plt.show()

# In the same plot
import matplotlib.pyplot as plt

plt.figure(figsize=(6,4))
ax1 = plt.gca()

# ===== 左轴：fT (GHz) =====
for i in range(len(L_set)):
    ax1.plot(gm_ID_vec, fT_vec[i]/1e9,
             label=f"fT (L={L_set[i]:.2f} µm)", linestyle='-')
ax1.set_xlabel("gm/ID (1/V)")
ax1.set_ylabel("fT (GHz)")
ax1.set_yscale("log")
ax1.tick_params(axis='y', labelcolor='tab:blue')
ax1.grid(True, which='both', ls=':')

# ===== 右轴：Av =====
ax2 = ax1.twinx()
for i in range(len(L_set)):
    ax2.plot(gm_ID_vec, Av_vec[i],
             linestyle='--', color=ax1.get_lines()[i].get_color(),
             label=f"Av (L={L_set[i]:.2f} µm)")
ax2.set_ylabel("Av (V/V)", color='tab:red')
ax2.tick_params(axis='y', labelcolor='tab:red')

# ===== 双图例（fT 实线 + Av 虚线）=====
lines_ft, labels_ft = ax1.get_legend_handles_labels()
lines_av, labels_av = ax2.get_legend_handles_labels()

first_legend = ax1.legend(lines_ft, labels_ft, loc='lower right', frameon=False)
ax2.legend(lines_av, labels_av, loc='upper left', frameon=False)
ax1.add_artist(first_legend)

# ===== 标题与布局 =====
plt.title("fT and Av vs gm/ID under different channel lengths")
plt.tight_layout()
plt.show()


```

- Conclusion from the former 2 plots
  - L larger -> A<sub>v</sub> bigger, because L larger -> channel length modulation smaller -> r<sub>o</sub> smaller
  - L larger -> f<sub>T</sub> smaller, because L larger -> c<sub>gg</sub> smaller & g<sub>m</sub> larger 
  - gm_ID larger -> A<sub>v</sub> smaller & f<sub>T</sub>

- **g<sub>m</sub>/I<sub>D</sub> systematic design methodology**
  - From design specifications and constrains, determine g<sub>m</sub>
  - Pick L: 
    - Small L -> High f<sub>T</sub>, small Area
    - Large L -> High A<sub>v</sub>, large area
  - Pick g<sub>m</sub>/I<sub>D</sub>: 
    - Large g<sub>m</sub>/I<sub>D</sub> -> Low-power, high A<sub>v</sub>
    - Small g<sub>m</sub>/I<sub>D</sub> -> High f<sub>T</sub>, small Area
  - Derive I<sub>D</sub> by using the result of g<sub>m</sub> and g<sub>m</sub>/I<sub>D</sub>
  - Determine W using I<sub>D</sub>/W denormalization

```python
# Set fixed parameters
gm_ID = 15
L = 0.1
fu = 1e9
CL = 1e-12
VDS = 0.55

# Direct Look up
fT = look_up_vs_gm_id(nch, "GM_CGG", gm_id = gm_ID, vds = VDS, l = L, vsb = 0)/(2*np.pi)
Av = look_up_vs_gm_id(nch, "GM_GDS", gm_id = gm_ID, vds = VDS, l = L, vsb = 0)
JD = look_up_vs_gm_id(nch, "ID_W", gm_id = gm_ID, vds = VDS, l = L, vsb = 0)
gm = 2*np.pi*fu*CL
ID = gm / gm_ID
W = ID / JD
VGS = look_up_vgs_vs_gm_id(nch, gm_id = gm_ID, l = L, vds = VDS, vsb = 0)

print("fT: ", fT)
print("Av: ", Av)
print("ID: ", ID)
print("JD: ", JD)
print("gm: ", gm)
print("W: ", W)
print("VGS: ", VGS)
print(nch.keys())
```
